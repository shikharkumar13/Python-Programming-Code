### Nested `if-else` statements

```python
marks = int(input('Enter your marks: '))

if marks >= 0 and marks <= 100:
    if marks >= 35 and marks <= 100:
        print('You Passed.')
    else:
        print('You failed')
else:
    print('Invalid Marks. Please mention a number between 0 and 100.')
```

### Real world example scenarios for conditional statements

#### Uber: Surge Pricing

Uber adjusts its prices dynamically based on how many people are requesting rides versus how many drivers are available. If demand is "High" (like during a rainstorm), they multiply the base fare by 2.0 to encourage more drivers to get on the road. If demand is "Medium" (like after a concert), they multiply it by 1.2. If demand is "Normal," the price stays the same.

```python
demand_level = "Medium"  # Options: Low, Normal, Medium, High
base_fare = 15.00        # Dollars

if demand_level == "High":
    surge_multiplier = 2.0
    print("Surge Pricing Active: High Demand")
elif demand_level == "Medium":
    surge_multiplier = 1.2
    print("Surge Pricing Active: Medium Demand")
else:
    surge_multiplier = 1.0
    print("Standard Pricing")

final_fare = base_fare * surge_multiplier
extra_cost = final_fare - base_fare

print(f"Status      : {demand_level}")
print(f"Surge       : {surge_multiplier}x")
print(f"Base Fare   : ₹{base_fare}")
print(f"Final Fare  : ₹{final_fare:.0f}")
print(f"Extra Cost  : ₹{extra_cost:.0f}")
print(f"Your final estimated fare is: ${final_fare:.2f}")
```

#### Amazon: Spending based Discount

Amazon rewards big spenders with bigger discounts. The discount tier is determined by the cart value, for ex. ₹10,000+ gets 20%, ₹5,000+ gets 15%, and so on. The program must calculate the exact savings and the final price to pay.

```python
cart_total = 3250   # Try: 300, 750, 2500, 6000, 12000

print(f"Your Cart Total: ₹{cart_total}")
print()

# Determine discount tier — checks from HIGHEST to LOWEST
if cart_total >= 10000:
    discount_percent = 20
    perk = "Free Priority Shipping!"
elif cart_total >= 5000:
    discount_percent = 15
    perk = "Free Standard Shipping!"
elif cart_total >= 2000:
    discount_percent = 10
    perk = "₹50 Shipping Discount!"
elif cart_total >= 500:
    discount_percent = 5
    perk = "Early Sale Access!"
else:
    # Cart is too small for any discount
    discount_percent = 0
    perk = "Add ₹500+ to unlock discounts!"

# Calculate the savings and final price
discount_amount = (cart_total * discount_percent) / 100
final_price = cart_total - discount_amount

# Display the results
print(f"Discount   : {discount_percent}%")
print(f"You Save   : ₹{discount_amount:.0f}")
print(f"Final Price: ₹{final_price:.0f}")
print(f"Perk       : {perk}")
```

#### Stripe: Fraud Detection 

Stripe processes millions of payments and needs to catch credit card fraud instantly. A simple rule-based flag checks for high-risk behavior. If a transaction attempt is greater than $10,000 AND the IP address of the computer making the purchase is in a different country than where the credit card was issued, the system must decline the transaction and flag it as "Suspicious."

```python
transaction_amount = 15000
card_country = "US"
ip_country = "RU"

if transaction_amount > 10000 and card_country != ip_country:
    transaction_status = "Declined"
    print("ALERT: Suspicious transaction detected. Account locked.")
else:
    transaction_status = "Approved"
    print("Transaction processed successfully.")
```

#### Swiggy: Discount Engine

Swiggy is running a marketing campaign: "Use code WELCOME50 for 50% off your first order (minimum order ₹199)." When a user enters the code, the system first checks if they are a new user. If they are, it then checks if their cart value is at least ₹199. We use nested if statements here so we can give the user a specific error message depending on exactly which rule they broke.

```python
is_new_user = True
cart_value = 150  # Rupees
promo_code_entered = "WELCOME50"

if promo_code_entered == "WELCOME50":
    if is_new_user:
        if cart_value >= 199:
            discount = cart_value * 0.50
            print(f"Promo applied! You saved ₹{discount}")
        else:
            print(f"Cart value too low. Add items worth ₹{199 - cart_value} to use this code.")
    else:
        print("Sorry, this promo code is for new users only.")
```

#### Uber: Driver Tracking (Condition-Based Updates)

When you book an Uber, the app shows a little car moving on your map. The app continuously requests the driver's GPS coordinates. It doesn't know how many GPS updates it will take to reach you—it just knows it needs to keep asking for updates while the trip status is "On Trip" and stop the moment the status changes to "Dropped."

```python
import time

trip_status = "On Trip"
distance_to_destination = 7   #kilometers

print("Trip started. Tracking driver location...\n")

while trip_status == "On Trip":
    print(f"You are {distance_to_destination}km away. Updating map...")
    distance_to_destination -= 1
    time.sleep(1)
    
    if distance_to_destination <= 0:
        trip_status = "Dropped"
        print("\n You have reached your destination.")
```

#### Netflix: Video Buffering

When you have a slow internet connection, Netflix pauses your movie to "buffer." It downloads small chunks of the video into your device's memory. The app uses a while loop that says: While the video buffer is less than 20% full, keep downloading chunks and show the spinning loading wheel. Once it hits 20%, stop the loop and resume playback.

```python
import time

buffer_percentage = 0
required_buffer = 20

print("Network connection slow. Pausing video to buffer...")

while buffer_percentage < required_buffer:
    buffer_percentage += 5 # Downloading chunks of data
    print(f"Buffering... {buffer_percentage}%")
    time.sleep(1)

print("\n Buffer threshold reached. Resuming playback.")
```

#### Swiggy: Delivery Status

After you order food on Swiggy, your app needs to know when the restaurant hands the food to the delivery partner. The app uses a while loop to "poll" (ask) the restaurant's system for an update every few seconds. It keeps asking while the order status is "Preparing." The loop stops the moment the status changes to "Picked Up."

```python
import time

order_status = "Preparing"
polling_attempts = 0

print("Order placed! Waiting for restaurant...\n")

while order_status != "Picked Up":
    polling_attempts += 1
    print(f"[Poll #{polling_attempts}] Checking status... Current status: {order_status}")
    time.sleep(1) # Wait a bit before asking again
    
    # Simulating the restaurant finishing the food on the 4th check
    if polling_attempts == 4:
        order_status = "Picked Up"
        print('Your order has been picked by the delivery partner.')

print("\n Your food is on the way!")
```

#### Instagram: Infinite Scroll (Event-Driven Loop)

Instagram doesn't load every single photo on the server when you open the app; that would crash your phone. Instead, it loads a few posts. Then, it uses a while loop to listen for your scrolling. While you are scrolling near the bottom of your feed, it fetches the next batch of posts. It stops fetching when you stop scrolling.

```python
user_is_scrolling = True
posts_loaded = 5

print(f"Initial load: {posts_loaded} posts on screen.")

while user_is_scrolling:
    print("User reached the bottom of the feed. Fetching more posts...")
    posts_loaded += 5
    print(f"Loading... Feed now has {posts_loaded} posts.")
    
    # Simulating the user stopping after 15 posts
    if posts_loaded >= 15:
        print("User stopped scrolling. Pausing background fetch.")
        user_is_scrolling = False
```

#### YouTube: Playlist Duration

When a user queues up a "Watch Later" playlist, YouTube needs to calculate the total runtime of all the videos combined. The system has a list of the individual video durations. To get the total, it needs to start with a timer at zero, loop through every video in the list, and add each video's length to the running total.

```python
# Video durations in minutes
video_durations = [12.5, 8.0, 22.3, 5.5, 14.1] 

# Initialize the accumulator variable
total_playlist_time = 0.0

print("Calculating playlist duration...")

# The loop goes through the list one by one
for duration in video_durations:
    total_playlist_time += duration
    print(f"Added video ({duration} min). Running total: {total_playlist_time:.1f} min")

print(f"\nTotal Watch Time: {total_playlist_time:.1f} minutes")
```

#### Facebook: Friend Suggestions

To suggest "People You May Know," Facebook looks at your network. One simple way to do this is to take a list of your best friend's connections and loop through it. The system checks if any of those people are already your friend. If they aren't, the system flags them as a potential new friend to suggest to you.

```python
my_friends = ["Alex", "Jordan", "Taylor"]
jordan_friends = ["Alex", "Sam", "Casey", "Taylor", "Riley"]

suggestions = []

print("Scanning for friend suggestions...\n")

for person in jordan_friends:
    # Check if the person is NOT me, and NOT already in my friend list
    if person not in my_friends and person != "Me":
        suggestions.append(person)

print(f"\nPeople You May Know: {suggestions}")
```

#### Amazon: Inventory Check 

Amazon warehouses hold millions of products. Every night, an automated system sweeps through the database to check stock levels. It loops through a list of products, and if it finds an item where the stock count has dropped to zero, it alerts the purchasing department to reorder that specific item.

```python
# A list of dictionaries representing warehouse inventory
inventory = [
    {"item": "Wireless Mouse", "stock": 45},
    {"item": "Mechanical Keyboard", "stock": 5},
    {"item": "USB-C Cable", "stock": 120},
    {"item": "Laptop Stand", "stock": 0}
]

print("Running daily inventory sweep...\n")

for product in inventory:
    # Access the 'stock' value for the current product
    if product["stock"] == 0:
        print(f"ALERT: {product['item']} is out of stock! Reorder immediately.")
    elif product["stock"] < 10:
        print(f"ALERT: {product['item']} is low on stock! Reorder immediately.")
```

#### Stripe: Monthly Statement (Categorizing and Summing)

At the end of the month, Stripe generates a financial statement for a business. It loops through all the transactions that occurred. However, not all transactions are the same—some are incoming payments, and some are outgoing refunds. The loop must look at the type of transaction, add payments to the total, and subtract refunds to calculate the final net revenue.

```python
# List of transaction amounts and their types
transactions = [
    {"amount": 150.00, "type": "payment"},
    {"amount": 45.00, "type": "refund"},
    {"amount": 200.00, "type": "payment"},
    {"amount": 10.00, "type": "refund"}
]

net_revenue = 0.0

for txn in transactions:
    if txn["type"] == "payment":
        net_revenue += txn["amount"]
        print(f"Processed Payment: +${txn['amount']:.2f}")
    elif txn["type"] == "refund":
        net_revenue -= txn["amount"]
        print(f"Processed Refund:  -${txn['amount']:.2f}")

print(f"\nMonthly Net Revenue: ${net_revenue:.2f}")
```
