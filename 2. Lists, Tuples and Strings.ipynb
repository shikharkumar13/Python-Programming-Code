### <font color = 'Green'>Some real world examples showing how lists can be used

#### Netflix: Chronological Watch History

**Scenario:** Netflix needs to track every show you watch to power the "Continue Watching" row. Because you might watch the same episode of The Office five times, duplicates must be allowed. The most recent view is usually appended to the end of the list.

```python
# The list represents shows watched from oldest to newest
watch_history = ["Stranger Things", "The Crown", "Dark", "The Office"]

# User watches a new show
watch_history.append("Black Mirror")

# User re-watches "The Office" (Duplicates are allowed)
watch_history.append("The Office")

# Finding the last watched show
last_watched = watch_history[-1]

print(f"Full History: {watch_history}")
print(f"Last Watched: {last_watched}")
```

#### Amazon: Shopping Cart Persistence

**Scenario:** An Amazon cart is a list of product IDs. Order matters because typically the items added last appear at the top. Amazon also needs to allow duplicates if a user clicks "Add to Cart" on the same item multiple times rather than changing the quantity field immediately.

```python
# Product IDs in the user's cart
amazon_cart = ["B07XJ8C8F5", "B08N5WRWJ5", "B07ZPKN6S2"]

# User adds another unit of the second item
amazon_cart.insert(0, "B08N5WRWJ5")

# User decides to remove the first item they added
amazon_cart.remove("B07XJ8C8F5")

print(f"Current Cart IDs: {amazon_cart}")
print(f"Total items in cart: {len(amazon_cart)}")
```

#### YouTube: Playlist Management

**Scenario:** A YouTube playlist is a classic list. You can reorder videos (move index 5 to index 0), delete videos, and add the same video to a playlist multiple times (e.g., a "Focus Study Music" playlist).

```python
# Video IDs in a "Python Tutorials" playlist
playlist = ["vid_101", "vid_102", "vid_103", "vid_104"]

# Move the 4th video (index 3) to the 1st position (index 0)
moving_vid = playlist.pop(3)
playlist.insert(0, moving_vid)

# Append a video at the very end
playlist.append("vid_999")

print(f"Custom Sorted Playlist: {playlist}")
```

#### Twitter/X: Post Drafts

**Scenario:** When you're composing multiple threads or saving ideas, Twitter stores these as a list of drafts. Since you might write several drafts with the exact same text (e.g., "Coming soon!"), a list is used to preserve every individual entry and the order in which they were created.

```python
# List of string drafts
drafts = [
    "Python is amazing!", 
    "Working on a new YouTube video...", 
    "Coming soon!"
]

# User deletes the oldest draft
del drafts[0]

# User edits the current latest draft
drafts[-1] = "Coming soon! Stay tuned for the Python series."

print(f"Your Saved Drafts: {drafts}")
```

### <font color = 'Green'>Some real world examples showing how lists can be used

#### Google Maps: Geographic Coordinates (Lat, Long)
**Scenario:** A specific landmark, like the Eiffel Tower, has a fixed physical location on Earth. If a developer accidentally changed the latitude in the code, the map would point to the wrong country. By storing these as a tuple, Google ensures the location remains a "source of truth."

```python
# Fixed coordinates for the Eiffel Tower (Latitude, Longitude)
eiffel_tower_coords = (48.8584, 2.2945)

# Attempting to change the latitude would raise a TypeError
#eiffel_tower_coords[0] = 49.0  # This will crash, protecting the data

print(f"Latitude: {eiffel_tower_coords[0]}")
print(f"Longitude: {eiffel_tower_coords[1]}")
```

### Stripe: Financial Transaction Snapshot

**Scenario:** Once a payment is processed, the core details—the amount, the currency, and the timestamp—must never change for auditing purposes. Stripe uses tuple-like structures to pass this "frozen" data between different microservices (e.g., from the payment gateway to the receipt generator).

```python
# A snapshot of a completed payment: (Amount, Currency, Date, TransactionID)
successful_payment = (29.99, "USD", "2024-05-20", "ch_3Oix22L")

# Unpacking the tuple for a receipt
amount, currency, date, txn_id = successful_payment

print(f"Receipt: {amount} {currency} processed on {date}")
```

#### Instagram: Hashtag & Mention Parsing

**Scenario:** When you post a caption, Instagram’s backend treats the entire caption as one long string. It uses string methods to "parse" (search) for characters like # or @ to turn them into clickable links.

```python
caption = "Learning #Python with the best community! @guido_van_rossum"

# Checking if a specific hashtag exists
has_python = "#Python" in caption

# Finding the position of a mention
mention_start = caption.find("@")
mention = caption[mention_start:] # Slicing the string

print(f"Hashtag Present: {has_python}")
print(f"User Mentioned: {mention}")
```

#### Google: URL Slugs (Sanitization)
**Scenario:** When you search for something, Google generates a URL. However, URLs cannot contain spaces. Google takes your search string and "sanitizes" it by replacing spaces with + or - characters to make it web-friendly.

```python
search_query = "How to learn Python in 2026"

# Sanitizing for a URL slug
url_slug = search_query.lower().replace(" ", "-")

print(f"https://www.google.com/search?q={url_slug}")
# Output: https://www.google.com/search?q=how-to-learn-python-in-2026
```

#### Stripe/Banking: Card Number Masking

**Scenario:** For security, PCI compliance rules forbid companies from showing your full 16-digit credit card number. Apps like Stripe or Amazon take your card string, keep the last 4 digits, and "mask" the rest with asterisks.

```python
card_number = "4532112233445566"

# Keeping only the last 4 characters
masked_card = "*" * 12 + card_number[-4:]

print(f"Payment Method: Visa ending in {masked_card}")
# Output: Payment Method: Visa ending in ************5566
```

#### Amazon: Dynamic Email Invoicing (f-strings)

**Scenario:** Amazon sends millions of emails daily. They don't write each one manually. They use f-strings (formatted strings) to inject variables (Name, Order ID, Price) into a pre-written text template.

```python
customer_name = "Abhishek"
order_item = "Sony Headphones"
delivery_date = "March 20"

# Using f-strings for clean output formatting
email_body = f"""
Hi {customer_name},
Your order for the '{order_item}' has been confirmed!
It will be delivered by {delivery_date}.
"""

print(email_body)
```

#### Netflix: Search Auto-Complete (Case Sensitivity)

**Scenario:** If you search for "STRANGER THINGS" or "stranger things," Netflix should show the same result. Developers use .lower() or .upper() to normalize the search string so that the comparison is "case-insensitive."

```python
movie_database = ["Stranger Things", "The Witcher", "Breaking Bad", "Pirates of the Caribbean: On Stranger Tides", ]
user_search = "stranger"

for movie in movie_database:
    # Converting both to lowercase for a fair comparison
    if user_search.lower() in movie.lower():
        print(f"Result found: {movie}")
```
